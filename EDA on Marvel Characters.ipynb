Handling Missing Data

```python
marvel.isna().sum()
```

```python
# page_id column irrelevant
marvel.drop('page_id',axis=1,inplace=True)
```

```python
# Lets drop column 'GSM' as most of its value are null so it isnot useful
marvel1 = marvel.drop('GSM',axis=1)
marvel1.head(5)
```

```python
# urlslug is an irrelevant column , also First Appearance and Year gives same meaning
marvel1.drop(['FIRST APPEARANCE','urlslug'],axis=1,inplace=True)
marvel1.head(4)
```

```python
# ID should always be available for each character
# gender of person should be known
marvel1.dropna(subset=['ID','SEX','APPEARANCES','Year','ALIGN'],inplace=True,ignore_index=True)
marvel1.isna().sum()
```

```python
# remaining null values in hair and eye , lets fill it
marvel1['EYE'] = marvel1['EYE'].ffill()
marvel1['HAIR'] = marvel1['HAIR'].bfill()
marvel1.isna().sum()
```

Encoding Categorical Features

```python
marvel2 = marvel1
marvel2
```

```python
# name can be encoded with label encoding
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
marvel2['name'] = label.fit_transform(marvel2['name'])

# or with this code - marvel2['name'].astype('category').cat.codes
```

```python
# onehotencoding for ID
from sklearn.preprocessing import OneHotEncoder

onehot = OneHotEncoder(sparse_output=False)
transformed = onehot.fit_transform(marvel2[['ID']])
features = onehot.get_feature_names_out()
df = pd.DataFrame(transformed,columns=features)
marvel3 = pd.concat([marvel2,df],axis=1)
marvel3.drop(['ID'],axis=1,inplace=True)
```

```python
# ordinal encoding for align
from sklearn.preprocessing import OrdinalEncoder

ord = OrdinalEncoder(categories=[['Bad Characters','Neutral Characters','Good Characters']])
val = ord.fit_transform(marvel3[['ALIGN']])
marvel3.drop('ALIGN',axis=1,inplace=True)
marvel3['ALIGN'] = val
```

```python
# ordinal encoding for ALIVE
from sklearn.preprocessing import OrdinalEncoder

ord = OrdinalEncoder(categories=[['Deceased Characters','Living Characters']])
val = ord.fit_transform(marvel3[['ALIVE']])
marvel3.drop('ALIVE',axis=1,inplace=True)
marvel3['ALIVE'] = val
```

```python
# label encoding
marvel3['EYE'] = marvel3['EYE'].astype('category').cat.codes
marvel3['HAIR'] = marvel3['HAIR'].astype('category').cat.codes
```

```python
# onehot
new = pd.get_dummies(data=marvel3['SEX']).astype(int)
marvel4 = pd.concat([marvel3,new],axis=1)
marvel4.drop('SEX',axis=1,inplace=True)
marvel4
```

```python
marvel4.drop('ID_Known to Authorities Identity',axis=1,inplace=True)
```

```python
marvel4.rename(columns={'Year':'Appearance Year',
                        'name':'Character Name',
                        'EYE':'Eye Color',
                        'HAIR':'Hair Color',
                        'ID_No Dual Identity':'Dual Identity',
                        'ID_Public Identity':'Public Identity',
                        'ID_Secret Identity':'Secret Identity',
                        'ALIGN':'Align',
                        'ALIVE':'Alive',
                        'Agender Characters':'Agender',
                        'Female Characters':'Female',
                        'Genderfluid Characters':'Genderfluid',
                        'Male Characters':'Male',
                        'APPEARANCES':'Appearances'
},inplace=True)
marvel4
```
